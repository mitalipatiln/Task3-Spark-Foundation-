```python
state_profit=df.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Profit')
state_profit[['Profit']].plot(kind="bar",figsize=(10,10))
plt.ylabel("Profit")
plt.show()
```

```python
state_profit=df.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Discount')
state_profit[['Discount']].plot(kind="bar",figsize=(15,4))
plt.ylabel("Discount")
plt.show()
```

```python
state_profit=df.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Sales')
state_profit[['Sales']].plot(kind="bar",figsize=(15,4))
plt.ylabel("Sales")
plt.show()
```

```python
state_profit=df.groupby(['State'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Profit')
state_profit[['Profit']].plot(kind="bar",figsize=(15,4),color="red")
plt.ylabel("Profit")
plt.show()
```

```python
state_profit=df.groupby(['State'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Discount')
state_profit[['Discount']].plot(kind="bar",figsize=(15,4),color="red")
plt.ylabel("Discount")
plt.show()
```

```python
state_profit=df.groupby(['State'])[['Profit','Discount','Sales']].mean()
state_profit=state_profit.sort_values('Sales')
state_profit[['Sales']].plot(kind="bar",figsize=(15,4),color="red")
plt.ylabel("Sales")
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
sns.distplot(df['Profit'],ax=axs[0][0],color="red")
sns.distplot(df['Discount'],ax=axs[0][1],color="blue")
sns.distplot(df['Sales'],ax=axs[1][0],color="yellow")
sns.distplot(df['Quantity'],ax=axs[1][1],color="green")
axs[0][0].set_title("Profit",fontsize=20)
axs[0][1].set_title("Discount",fontsize=20)
axs[1][0].set_title("Sales",fontsize=20)
axs[1][1].set_title("Quantity",fontsize=20)
plt.tight_layout()
```
